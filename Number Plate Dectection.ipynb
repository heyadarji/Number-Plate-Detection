# (1) Create a neural network model using Keras for classifying diabetes from the pima-indians-diabetes dataset. Compare the accuracy of the model using different activation functions like Relu, Leaku Relu and Tanh. 


```python
from numpy import loadtxt
from keras.models import Sequential
from keras.layers import Dense
```

```python
dataset = loadtxt('pima-indians-diabetes.csv', delimiter=',')
dataset
```

```python
# split into input (X) and output (y) variables
X = dataset[:,0:8]
y = dataset[:,8]
```

```python
model = Sequential()
model.add(Dense(12, input_dim=8, activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.fit(X, y, epochs=150, batch_size=10)
```

```python
# evaluate the keras model
_, accuracy = model.evaluate(X, y)
print('Accuracy: %.2f' % (accuracy*100))
```

# (2)Create a model for number plate identification through an image and highlight the number plate using red coloured rectangle with thick edges. 

```python
import cv2
#load the cascade
np_cascade=cv2.CascadeClassifier('np.xml')
#read the image
img =cv2.imread('1.jpg')
#convert color image into grayscale
grey=cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
#detect Plate
np=np_cascade.detectMultiScale (grey, 1.04, 4)
#draw rectengal on Plate

for (x, y, w, h) in np:
    cv2.rectangle(img, (x, y), (x+w , y+h), (20, 100 ,180), 2)
# cv2.imshow(img)
cv2.imshow('photo',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
